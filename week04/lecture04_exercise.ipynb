**Rules:**
1. Colour used **sparingly** — one categorical variable, no rainbow

3. `size_max` set when using bubble size
4. Log scale for GDP per capita
5. Insight title

```python
import pandas as pd
import plotly.express as px


# Dataset: Gapminder — GDP, Life Expectancy, Population by Country
# Source: Gapminder Foundation (gapminder.org)

df = px.data.gapminder()
print(f"Loaded: {len(df)} rows")
print(df.head())
```

```python
# explore

print(df.info())
print("Years:", sorted(df['year'].unique()))
print("Continents:", df['continent'].unique())
print(df.describe().round(1))
```

## Task 1 — Scatter: life expectancy change over time

**What to build:** A scatter showing **GDP per capita vs life expectancy** for **two years** (2002 and 2007) to show how both moved — use **colour for year** (just 2 colours), **one continent only**.

Choose any continent except Africa (that was the example). Highlight the change direction.

> 💡 Filter: `df.loc[df['continent'] == 'YOUR_CHOICE']` then filter years


```python
# Task 1
# Filter Europe data for 2002 and 2007
scatter_df = df[
    (df['continent'] == 'Europe') &
    (df['year'].isin([2002, 2007]))
]

# Create scatter plot
fig = px.scatter(
    scatter_df,
    x='gdpPercap',
    y='lifeExp',
    color='year',
    hover_name='country',
    log_x=True,
    title='European countries improved life expectancy as GDP increased',
    labels={
        'gdpPercap': 'GDP per Capita',
        'lifeExp': 'Life Expectancy'
    },
    template='simple_white'
)

fig.show()

# Save chart
fig.write_image("task1_chart.png")
```

## Task 2 — Bubble chart: tell a story

**What to build:** A bubble chart (full 2007 dataset, all countries) where:
- x = GDP per capita (log scale)
- y = life expectancy
- size = population
- colour = ONE continent highlighted (your choice), all others grey
- At least one annotation explaining the highlighted group's story

> This is the grey-and-highlight technique applied to a bubble chart.


```python
# Task 2
# Filter 2007 data
bubble_df = df[df['year'] == 2007]

# Highlight Asia
highlight_continent = 'Asia'

# Create color column
bubble_df['highlight'] = bubble_df['continent'].apply(
    lambda x: highlight_continent if x == highlight_continent else 'Other'
)

# Bubble chart
fig = px.scatter(
    bubble_df,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    color='highlight',
    hover_name='country',
    log_x=True,
    color_discrete_map={
        'Asia': 'red',
        'Other': 'lightgrey'
    },
    title='Asian countries combine large populations with rising life expectancy',
    labels={
        'gdpPercap': 'GDP per Capita',
        'lifeExp': 'Life Expectancy'
    },
    template='simple_white'
)

# Add annotation
fig.add_annotation(
    x=10000,
    y=75,
    text='Many Asian countries achieved higher life expectancy with growing economies',
    showarrow=True,
    arrowhead=2
)

fig.show()

# Save chart
fig.write_image("task2_chart.png")
```
